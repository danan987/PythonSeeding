## Iterating through pandas dataframes
- We can use the DataFrame method iterrows() to create a list of the DataFrame rows that we access through variables

```python
import pandas as pd
```

```python
google_drive_url = "https://docs.google.com/spreadsheets/d/10QIbM1dUJpi6RRPMmQb2F_URgg-dHkFi/edit?usp=sharing&ouid=117218510125754224635&rtpof=true&sd=true"
raw_url = 'https://drive.google.com/uc?id=' + google_drive_url.split('/')[-2]
```

```python
sheet_names = ["Customers",
               "Employees",
               "Managers",
              ]
```

```python
data = pd.read_excel(raw_url, sheet_name = sheet_names)
```

```python
data["Managers"]
```

```python
seedusers_string = ""

# Add Customers
for customer in data["Customers"].to_dict(orient="records"):
    seedusers_string += f"""
        AllUsers.Add(new AddUserModel()
        {{
            User = new AppUser()
            {{
                FirstName = "{customer['First Name']}",
                LastName = "{customer['Last Name']}",
                Email = "{customer['Email']}",
                Address = "{customer['Street']}",
                Birthday = DateTime.Parse("{customer['Birthday']}"),  // assuming 'Birthday' is in a compatible format
                UserName = "{customer['Email']}",  // using Email as UserName
            }},
            Password = "{customer['Password']}",
            RoleName = "Customer"
        }});
"""

# Add Employees
for employee in data["Employees"].to_dict(orient="records"):
    seedusers_string += f"""
        AllUsers.Add(new AddUserModel()
        {{
            User = new AppUser()
            {{
                FirstName = "{employee['First Name']}",
                LastName = "{employee['Last Name']}",
                Email = "{employee['User Name']}",
                UserName = "{employee['User Name']}",  // using Email as UserName
                IsHired = true,
                Address = " ",
                Birthday = DateTime.Parse("2002-05-02 00:00:00"),
                PhoneNumber = "(512)555-1234"
            }},
            Password = "{employee['Password']}",
            RoleName = "Employee"
        }});
"""

# Add Managers
for manager in data["Managers"].to_dict(orient="records"):
    seedusers_string += f"""
        AllUsers.Add(new AddUserModel()
        {{
            User = new AppUser()
            {{
                FirstName = "{manager['First Name']}",
                LastName = "{manager['Last Name']}",
                Email = "{manager['Email Address']}",
                UserName = "{manager['Email Address']}",  // using Email as UserName
                IsHired = true,
                Address = " ",
                Birthday = DateTime.Parse("2002-05-02 00:00:00"),
                PhoneNumber = "(512)555-1234"
            }},
            Password = "{manager['Password']}",
            RoleName = "Manager"
        }});
"""
```

```python
print(seedusers_string)
```

```python
file = open("SeedUsers", "w")
file.write(seedusers_string)
file.close()
```
