## Iterating through pandas dataframes
- We can use the DataFrame method iterrows() to create a list of the DataFrame rows that we access through variables

```python
import pandas as pd
```

```python
google_drive_url = "https://docs.google.com/spreadsheets/d/10QIbM1dUJpi6RRPMmQb2F_URgg-dHkFi/edit?usp=sharing&ouid=117218510125754224635&rtpof=true&sd=true"
raw_url = 'https://drive.google.com/uc?id=' + google_drive_url.split('/')[-2]
```

```python
sheet_names = ["Schedule","Movies"]
```

```python
data = pd.read_excel(raw_url, sheet_name = sheet_names)
print(data.keys())
```

```python
ticket_data = {
    'TicketPriceID': [1, 2, 3, 4, 5],
    'TicketType': ['Special Event', 'Matinee', 'Discount Tuesday', 'Weekday (Base)', 'Weekends'],
}

tickets = pd.DataFrame(ticket_data)

# Create a dictionary to map genre names to GenreID
ticket_mapping = dict(zip(tickets['TicketType'], tickets['TicketPriceID']))
movie_id_mapping = {movie["Title"]: movie["MovieID"] for movie in data["Movies"].to_dict(orient="records")}

```

```python
schedules_string = ""
schedule_id = 1  # Starting ID for schedules

for i, schedule in enumerate(data["Schedule"].to_dict(orient="records")):
    # Extracting date and time information
    date_time = schedule["DateTime"]
    year = date_time.year
    month = date_time.month
    day = date_time.day
    hour = date_time.hour
    minute = date_time.minute

    # Determine if it's a special event
    is_special_event = 'false' if pd.isna(schedule["Special Event?"]) or schedule["Special Event?"].lower() != 'yes' else 'true'

    # Assigning 'MovieName' from the schedule
    MovieName = schedule["Movie"]
    movie_id = movie_id_mapping.get(MovieName)



    # Fetch MovieID based on MovieName
# Create a dictionary to store the mapping of MovieTitle to MovieID
    TheaterType = "One" if schedule["Theater"] == "Theater 1" else "Two"
    ticketpriceid = ticket_mapping.get(schedule["Price"])

    price = schedule["Price"]
    new_movie_id = f"{int(movie_id) + 3000}"


    # Construct the Schedule object
    schedules_string += f"""AllScheduleSlots.Add(new ScheduleSlot()
        {{
            Movie = context.Movies.SingleOrDefault( m => m.MovieID == {new_movie_id}),
            StartTime =  new DateTime({year}, {month}, {day}, {hour}, {minute}, 0),
            EndTime =  new DateTime({year}, {month}, {day}, {hour}, {minute}, 0).AddMinutes(context.Movies.SingleOrDefault( m => m.MovieID =={new_movie_id}).RunTime?? 0),
            TheatreType = TheatreType.{TheaterType},
            TicketPrice =  context.TicketPrices.SingleOrDefault(g => g.TicketPriceID == {ticketpriceid}),
            IsSpecialEvent = {is_special_event},
            IsPublished = true
        }});
    """
    schedule_id += 1
```

```python
print(schedules_string)
```

```python
file = open("SeedScheduleSlot", "w")
file.write(schedules_string)
file.close()
```
